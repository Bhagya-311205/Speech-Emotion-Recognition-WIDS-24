## Matplotlib Assignment

This assignment will teach you how to create various types of visualizations, such as line plots, bar charts, histograms, and scatter plots, to analyze and present data effectively.

*The matplotlib library is already imported for you.*

```python
import matplotlib.pyplot as plt
```

Now, using the data loaded in the previous section (or by reloading it), create various plots based on the `df` DataFrame.

- Read Total profit of all months and show it using a line plot. <br>
Generated line plot must include the following properties:â€“

    - Line Style dotted and Line-color should be red
    - Show legend at the lower right location.
    - X label name = Month Number
    - Y label name = Profit in dollar
    - Add a circle marker.
    - Line marker color as read
    - Line width should be 3

```python
# your code here
plt.plot(df['month_number'],df['total_profit'], linestyle='dotted', color='red',marker='o', markerfacecolor='red', linewidth=3)
plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.show()
```

- Read all product sales data and show it using a multiline plot. Display the number of units sold per month for each product using multiline plots. (i.e., separate plotline and different color for each product ).

```python
# your code here
plt.plot(df['month_number'],df['facecream'], label='facecream')
plt.plot(df['month_number'],df['facewash'], label='facewash')
plt.plot(df['month_number'],df['toothpaste'], label='toothpaste')
plt.plot(df['month_number'],df['bathingsoap'], label='bathingsoap')
plt.plot(df['month_number'],df['shampoo'], label='shampoo')
plt.plot(df['month_number'],df['moisturizer'], label='moisturizer')
plt.legend()
plt.show()
```

- Calculate total sale data for last year for each product and show it using a Pie chart. <br>
Note: In Pie chart display Number of units sold per year for each product in percentage.

```python
# your code here
labels = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
sizes = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

- Read total units sold per month data and show it using the Bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
# your code here
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
months = df['month_number']
bar_width = 0.1
positions = range(len(months))
for i, product in enumerate(products):
    plt.bar([p + i * bar_width for p in positions], df[product], width=bar_width, label=product)
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.title('Units Sold Per Month for Each Product')
plt.xticks([p + (len(products) - 1) * bar_width / 2 for p in positions], months)
plt.legend()
plt.show()
```
